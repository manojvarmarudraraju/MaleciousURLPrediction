```python
Post_x_0_valid= len(df[(df['Valid url']==1 ) & (df['Global rank']==0)])/20
Post_x_1_valid= len(df[(df['Valid url']==1 ) & (df['Global rank']==1)])/20
Post_x_0_invalid= len(df[(df['Valid url']==0 ) & (df['Global rank']==0)])/20
Post_x_1_invalid= len(df[(df['Valid url']==0 ) & (df['Global rank']==1)])/20

Post_y_0_valid= len(df[(df['Valid url']==1 ) & (df['Alexa USA rank']==0)])/20
Post_y_1_valid= len(df[(df['Valid url']==1 ) & (df['Alexa USA rank']==1)])/20
Post_y_0_invalid= len(df[(df['Valid url']==0 ) & (df['Alexa USA rank']==0)])/20
Post_y_1_invalid= len(df[(df['Valid url']==0 ) & (df['Alexa USA rank']==1)])/20

Post_z_0_valid= len(df[(df['Valid url']==1 ) & (df['cPR score out of 10']==0)])/20
Post_z_1_valid= len(df[(df['Valid url']==1 ) & (df['cPR score out of 10']==1)])/20
Post_z_0_invalid= len(df[(df['Valid url']==0 ) & (df['cPR score out of 10']==0)])/20
Post_z_1_invalid= len(df[(df['Valid url']==0 ) & (df['cPR score out of 10']==1)])/20

if(Post_x_0_valid==0):
    Post_x_0_valid+=0.01
    Post_x_1_valid-=0.01
if(Post_y_0_valid==0):
    Post_y_0_valid+=0.01
    Post_y_1_valid-=0.01
if(Post_z_0_valid==0):
    Post_z_0_valid+=0.01
    Post_z_1_valid-=0.01
    

mydict = {
    'Post_x_0_valid':Post_x_0_valid,
    'Post_x_1_valid':Post_x_1_valid,
    'Post_x_0_invalid':Post_x_0_invalid,
    'Post_x_1_invalid':Post_x_1_invalid,
    
    'Post_y_0_valid':Post_y_0_valid,
    'Post_y_1_valid':Post_y_1_valid,
    'Post_y_0_invalid':Post_y_0_invalid,
    'Post_y_1_invalid':Post_y_1_invalid,
    
    'Post_z_0_valid':Post_z_0_valid,
    'Post_z_1_valid':Post_z_1_valid,
    'Post_z_0_invalid':Post_z_0_invalid,
    'Post_z_1_invalid':Post_z_1_invalid
    
}
```

```python
a=0
b=1
c=1
stra1="Post_x_"+str(a)+"_valid"
strb1="Post_y_"+str(b)+"_valid"
strc1="Post_z_"+str(c)+"_valid"

stra0="Post_x_"+str(a)+"_invalid"
strb0="Post_y_"+str(b)+"_invalid"
strc0="Post_z_"+str(c)+"_invalid"

prob_valid=0.5
prob_invalid=0.5
```

```python
prob_valid_given_abc=(prob_valid*mydict[stra1]*mydict[strb1]*mydict[strc1])/((prob_valid*mydict[stra1]*mydict[strb1]*mydict[strc1])+(prob_invalid*mydict[stra0]*mydict[strb0]*mydict[strc0]))
prob_invalid_given_abc=(prob_invalid*mydict[stra0]*mydict[strb0]*mydict[strc0])/((prob_valid*mydict[stra1]*mydict[strb1]*mydict[strc1])+(prob_invalid*mydict[stra0]*mydict[strb0]*mydict[strc0]))
print(prob_valid_given_abc,prob_invalid_given_abc)
```
